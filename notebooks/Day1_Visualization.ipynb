```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

Moving spines and add legend: 

```python
# create new figure: 
plt.figure(figsize=(10,6), dpi=100)
# create new subplot from a 1x1 grid
plt.subplot(111)
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)
# plot cosine in blue with continuous line of width 1 px
plt.plot(X,C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
# plot sine in green
plt.plot(X,S,color="red", linewidth=2.5, linestyle="-", label="sine")
# set x limints
plt.xlim(-4.0,4.0)
# set x ticks
plt.xticks([-np.pi, -np.pi/2,0,np.pi/2,np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
# set y limits
plt.ylim(-1.0,1.0)
# set y ticks
plt.yticks([-1,0,+1],
           [r'$-1$', r'$0$', r'$+1$'])

# moving spines
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data",0))

# legend
plt.legend(loc="upper left", frameon=False)

plt.show()
```

Add annotation to some points:

```python
# create new figure: 
plt.figure(figsize=(10,6), dpi=100)
# create new subplot from a 1x1 grid
plt.subplot(111)
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)
# plot cosine in blue with continuous line of width 1 px
plt.plot(X,C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
# plot sine in green
plt.plot(X,S,color="red", linewidth=2.5, linestyle="-", label="sine")
# set x limints
plt.xlim(-4.0,4.0)
# set x ticks
plt.xticks([-np.pi, -np.pi/2,0,np.pi/2,np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
# set y limits
plt.ylim(-1.0,1.0)
# set y ticks
plt.yticks([-1,0,+1],
           [r'$-1$', r'$0$', r'$+1$'])

# moving spines
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data",0))


# legend
plt.legend(loc="upper left", frameon=False)

# Annotation
t = 2*np.pi/3
# Add dashed lines
plt.plot([t,t],[0,np.cos(t)], color="blue", linewidth=1.5,linestyle="--")
plt.plot([t,t],[0,np.sin(t)], color="red", linewidth=1.5,linestyle="--")
# add points
plt.scatter([t,],[np.cos(t),], 50, color="blue")
plt.scatter([t,],[np.sin(t),], 50, color="red")
# Label points
plt.annotate(r'$\cos\left(\frac{2\pi}{3}\right)=\frac{-1}{2}$', 
            xy=(t,np.cos(t)),xycoords="data", 
            xytext=(-90,-50), textcoords="offset points",fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
            )
plt.annotate(r'$\sin\left(\frac{2\pi}{3}\right)=\frac{\sqrt{3}}{2}$', 
            xy=(t,np.sin(t)),xycoords="data", 
             xytext=(+10,+30), textcoords="offset points",fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")
            )
# highlight tick labels

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white",edgecolor="None", alpha=0.65))
plt.show()
```

# Plot the Normal Distribution
Now you should be able to reproduce the following plot (taken from Wikipedia):
<img src='https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/Normal_Distribution_PDF.svg/720px-Normal_Distribution_PDF.svg.png'>

```python
plt.figure(figsize=(10,6), dpi=100)
# create new subplot from a 1x1 grid
plt.subplot(111)
plt.xlim(-5.5,5.5)

x = np.linspace(-5, 5, 100)
plt.plot(x,mlab.normpdf(x, 0,math.sqrt(0.2)),color="blue", linewidth=2.5, linestyle="-",
         label=r'$\mu=0, \,\,\,\,\,\,\,\,\,\sigma^2=0.2$')
plt.plot(x,mlab.normpdf(x, 0,math.sqrt(1.0)),color="red", linewidth=2.5, linestyle="-",
         label=r'$\mu=0, \,\,\,\,\,\,\,\,\,\sigma^2=1.0$'), 
plt.plot(x,mlab.normpdf(x, 0,math.sqrt(5.0)),color="orange", linewidth=2.5, linestyle="-",
         label=r'$\mu=0, \,\,\,\,\,\,\,\,\,\sigma^2=5.0$')
plt.plot(x,mlab.normpdf(x, -2,math.sqrt(0.5)),color="green", linewidth=2.5, linestyle="-",
         label=r'$\mu=-2, \, \sigma^2=0.5$')

plt.xticks(range(-5,6), fontsize=14)
plt.yticks(np.arange(0,1.2,.2), fontsize=14)
plt.xlabel(r'$x$', fontsize=25)
plt.ylabel(r'$\varphi\mu,\sigma^2(x)$', fontsize=25)


ax = plt.gca()
ax.set_axis_bgcolor("white")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")


ax.spines["right"].set_color("black")
ax.spines["top"].set_color("black")

#ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position("left")


# legend
plt.legend(loc="upper right", frameon=True, 
           fontsize=14,markerfirst=False).get_frame().set_color('white')
plt.grid(True,color="lightgray")
plt.show()
```
